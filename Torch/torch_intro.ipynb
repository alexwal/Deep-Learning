```lua
require 'nn'
require 'optim'
Plot = require 'itorch.Plot'
```

```lua
-- DATA

function underlying(x)
    return torch.pow(1 + torch.exp((-x + 12)/100.), 2)
end

train_data, train_labels = {}, {}
validation_data, validation_labels = {}, {}
N = 1000
validation_percentage = 0.3
validation_N = math.floor(validation_percentage * N)
train_N = N - validation_N
min, max = -100, 100

for i = 1, validation_N do
    local x = torch.uniform(min - 10, max + 10)
    local y = underlying(x)
    table.insert(validation_data, torch.Tensor{x})
    table.insert(validation_labels, torch.Tensor{y})
end

for i = validation_N + 1, N do
    local x = torch.uniform(min, max)
    local y = underlying(x)
    table.insert(train_data, torch.Tensor{x})
    table.insert(train_labels, torch.Tensor{y})
end

validation_data = torch.cat(validation_data):resize(#validation_data, 1)
validation_labels = torch.cat(validation_labels):resize(#validation_labels, 1)
train_data = torch.cat(train_data):resize(#train_data, 1)
train_labels = torch.cat(train_labels):resize(#train_labels, 1)

-- PREPROCESS (optionally)
-- train_data_mean = train_data:mean()
-- train_data = train_data - train_data_mean
-- validation_data = validation_data - train_data_mean
```

```lua
-- SETUP 

-- Define sizes and model architecture.
inputs, hidden1_units, hidden2_units, hidden3_units, hidden4_units, outputs = 1, 3, 2, 2, 2, 1

model = nn.Sequential()
model:add(nn.Linear(inputs, hidden1_units))
model:add(nn.ReLU())
model:add(nn.Linear(hidden1_units, hidden2_units))
model:add(nn.Dropout(0.1))
model:add(nn.ReLU())
model:add(nn.Linear(hidden2_units, hidden3_units))
model:add(nn.Sigmoid())
model:add(nn.Linear(hidden3_units, hidden4_units))
model:add(nn.Tanh())
for i = 1, 4 do
    model:add(nn.Linear(hidden4_units, hidden4_units))
    model:add(nn.ReLU())
end
model:add(nn.Linear(hidden4_units, outputs))

criterion = nn.MSECriterion()

print(model)
```

```lua
-- TRAIN

epochs = 2400
check_every = 20
print_every = 100
batchSize = 32
learning_rate = 1e-3

validation_losses, train_losses = {}, {}

-- Use optim's SGD algorithm to train the model.
-- We'll need to provide its LR via an optimization state table.
optimState = {learningRate = learning_rate}

model:training()

-- The optim module expects the model parameters being optimized (weights + biases)
-- AND their gradients to be stored as 1D Tensors. We can get a nice 1D-
-- View of our params and gradParams with a simple call to model:getParameters().
-- These will become our new model params; old references to model params now invalid.
params, gradParams = model:getParameters()

for epoch = 1, epochs do
    -- Shuffle input data order
    local shuffle = torch.randperm(train_N)
    -- Train on all train_data batch by batch:
    for i, batch_inds in ipairs(shuffle:split(batchSize, 1)) do
        -- Process a single batch
        batchData = train_data:index(1, batch_inds:long()) -- dim, LongTensor{inds}
        batchLabels = train_labels:index(1, batch_inds:long())
        -- Closure that computes loss and gradParams, needed for optim:
        local function feval(params)
            -- Input: params = current model weights.
            gradParams:zero() -- Reset accumulation of grads to compute gradient only on this batch.
            local outputs = model:forward(batchData)
            local loss = criterion:forward(outputs, batchLabels) -- By default, the losses are averaged over
                                                                 -- observations for each minibatch. However, if
                                                                 -- the field sizeAverage is set to false, the
                                                                 -- losses are instead summed.
            local dloss_doutput = criterion:backward(outputs, batchLabels)
            model:backward(batchData, dloss_doutput) -- performs backpropagation pass, computing gradParams.
                                                     -- dloss_douput * doutput_dweights = dloss_dweights [gradParams]
            return loss, gradParams
        end
        optim.sgd(feval, params, optimState)
    end
    if epoch % check_every == 0 then
        model:evaluate()
        local validation_loss = criterion:forward(model:forward(validation_data), validation_labels)
        local train_loss = criterion:forward(model:forward(train_data), train_labels)
        table.insert(validation_losses, validation_loss)
        table.insert(train_losses, train_loss)
        model:training()
    end
    if epoch % print_every == 0 then
        print("On epoch", epoch, "of", epochs)
    end
end

-- EVALUATE

model:evaluate()
loss = criterion:forward(model:forward(validation_data), validation_labels)
print("Final validation loss:", loss)
loss = criterion:forward(model:forward(train_data), train_labels)
print("Final train loss:", loss)

-- PLOT

local t = check_every * torch.range(1, #train_losses)
local plot = Plot()
plot:line(t, train_losses,      'red',  'Train')
plot:line(t, validation_losses, 'blue', 'Validation')
plot:xaxis('Epoch')
plot:legend(true):title('Loss over time')
plot:draw()
```
